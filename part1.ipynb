```python
import os
import cv2
from yt_dlp import YoutubeDL

def download_youtube_video(url, save_path="downloaded_video.mp4"):
    ydl_opts = {
        'format': 'bestvideo+bestaudio/best',
        'outtmpl': save_path,
        'noplaylist': True,
    }
    with YoutubeDL(ydl_opts) as ydl:
        print(f"Downloading video from: {url}")
        ydl.download([url])
    print(f"Video downloaded to: {save_path}")

def extract_frames(video_path, output_dir, frame_interval=1):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps == 0:
        print("Warning: FPS is 0. Setting default FPS to 30.")
        fps = 30

    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Total Frames: {frame_count}, FPS: {fps}")

    frame_number = 0
    success, image = cap.read()
    if not success:
        print("Error: Unable to read the video file.")
        return

    while success:
        time_in_seconds = frame_number / fps
        if time_in_seconds.is_integer() and time_in_seconds % frame_interval == 0:
            frame_time = int(time_in_seconds)
            frame_filename = os.path.join(output_dir, f"frame_{frame_time:04d}.jpg")
            cv2.imwrite(frame_filename, image)
            print(f"Saved frame: {frame_filename}")
        success, image = cap.read()
        frame_number += 1

    cap.release()
    print(f"Frames saved to: {output_dir}")

# YouTube URL 및 경로 설정
youtube_url = "https://www.youtube.com/watch?v=OxnSjJVkoCs"
video_path = "downloaded_video.mp4"
output_dir = "frames"

# Step 1: 동영상 다운로드
download_youtube_video(youtube_url, save_path=video_path)

# Step 2: ffmpeg로 파일 재인코딩 (필요시)
if os.path.exists(video_path):
    !ffmpeg -i downloaded_video.mp4 -vcodec libx264 fixed_video.mp4
    video_path = "fixed_video.mp4"

# Step 3: 프레임 추출
extract_frames(video_path, output_dir, frame_interval=1)
```

```python
def extract_all_frames(video_path, output_dir):
    """
    Extracts all frames from a video file and saves them as images.
    """
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"Total Frames: {frame_count}, FPS: {fps}")

    frame_number = 0
    success, image = cap.read()
    if not success:
        print("Error: Unable to read the video file.")
        return

    while success:
        # 각 프레임 저장
        frame_filename = os.path.join(output_dir, f"frame_{frame_number:05d}.jpg")
        cv2.imwrite(frame_filename, image)
        print(f"Saved frame: {frame_filename}")

        success, image = cap.read()
        frame_number += 1

    cap.release()
    print(f"All frames saved to: {output_dir}")
```

```python
video_path = "fixed_video.mp4"  # 동영상 파일 경로
output_dir = "all_frames"      # 저장 폴더 이름

# 모든 프레임 저장 실행
extract_all_frames(video_path, output_dir)
```
